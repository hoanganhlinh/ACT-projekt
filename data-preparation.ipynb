## Przygotowanie danych

### 1. Zmiana nazw kolumn

```python
import re

def przeksztalc_nazwe(nazwa):
    # Usuń nawiasy ()
    nazwa = re.sub(r"[()]", "", nazwa)
    # Zamień wszystkie nielitery na _
    nazwa = re.sub(r"[^a-zA-Z]", "_", nazwa)
    # Zamień wielokrotne podłogi na jedną
    nazwa = re.sub(r"_+", "_", nazwa)
    # Usuń podłogi z początku i końca
    nazwa = nazwa.strip("_")
    # Zamień na lowercase
    return nazwa.lower()

# Zmień wszystkie kolumny w DataFrame
df_metabric.columns = [przeksztalc_nazwe(col) for col in df_metabric.columns]
```

### 6. Kategoryzacja zmiennych ciągłych

```python
import pandas as pd
import numpy as np

df_metabric['age_c'] = np.select(
    [
        df_metabric['age_at_diagnosis'] < 49,
        (df_metabric['age_at_diagnosis'] >= 49) & (df_metabric['age_at_diagnosis'] < 59),
        (df_metabric['age_at_diagnosis'] >= 59) & (df_metabric['age_at_diagnosis'] < 68),
        df_metabric['age_at_diagnosis'] >= 68
    ],
    [1, 2, 3, 4],
    default=np.nan
)

df_metabric['ts_c'] = np.select(
    [
        df_metabric['tumor_size'] == 0,
        (df_metabric['tumor_size'] > 0) & (df_metabric['tumor_size'] < 5),
        (df_metabric['tumor_size'] >= 5) & (df_metabric['tumor_size'] < 10),
        (df_metabric['tumor_size'] >= 10) & (df_metabric['tumor_size'] < 20),
        (df_metabric['tumor_size'] >= 20) & (df_metabric['tumor_size'] < 50),
        (df_metabric['tumor_size'] >= 50) & (df_metabric['tumor_size'] < 80),
        df_metabric['tumor_size'] >= 80
    ],
    [0, 1, 2, 3, 4, 5, 6],
    default=np.nan
)

df_metabric['npi_c'] = np.select(
    [
        df_metabric['nottingham_prognostic_index'] < 1,
        (df_metabric['nottingham_prognostic_index'] >= 1) & (df_metabric['nottingham_prognostic_index'] < 2),
        (df_metabric['nottingham_prognostic_index'] >= 2) & (df_metabric['nottingham_prognostic_index'] < 3),
        (df_metabric['nottingham_prognostic_index'] >= 3) & (df_metabric['nottingham_prognostic_index'] < 4),
        (df_metabric['nottingham_prognostic_index'] >= 4) & (df_metabric['nottingham_prognostic_index'] < 5),
        (df_metabric['nottingham_prognostic_index'] >= 5) & (df_metabric['nottingham_prognostic_index'] < 6),
        df_metabric['nottingham_prognostic_index'] >= 6
    ],
    [1, 2, 3, 4, 5, 6, 7],
    default=np.nan
)

df_metabric['lnep_c'] = np.select(
    [
        df_metabric['lymph_nodes_examined_positive'] == 0,
        (df_metabric['lymph_nodes_examined_positive'] > 0) & (df_metabric['lymph_nodes_examined_positive'] <= 2),
        (df_metabric['lymph_nodes_examined_positive'] > 2) & (df_metabric['lymph_nodes_examined_positive'] <= 6),
        (df_metabric['lymph_nodes_examined_positive'] > 6) & (df_metabric['lymph_nodes_examined_positive'] <= 15),
        df_metabric['lymph_nodes_examined_positive'] > 15
    ],
    [0, 1, 2, 3, 4],
    default=np.nan
)

df_metabric = df_metabric.drop(columns=['age_at_diagnosis', 'tumor_size', 'nottingham_prognostic_index', 'lymph_nodes_examined_positive'])
```

### TABLICE KONTYNGENCJI

```python
def count_censoring_separately(df, varlist, censor_col='c'):
    tables = {}
    for var in varlist:
        ct = pd.crosstab(df[var], df[censor_col])
        ct.columns = ['not_censored (c=0)', 'censored (c=1)']  # jeśli masz 0 i 1 w c
        tables[var] = ct
    return tables

# Przykład użycia:
varlist = ['lnep_c', 'gene_classifier_subtype', 'age_c', 'radio_therapy', 'ts_c', 'hormone_therapy', 'her_status']

tables_dict = count_censoring_separately(df_metabric, varlist)

cols = ['lnep_c', 'gene_classifier_subtype', 'age_c', 'radio_therapy', 'ts_c', 'hormone_therapy', 'her_status']
for col in cols:
    print(tables_dict[col])
```
